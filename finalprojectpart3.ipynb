```python
# STEP 3: Residual analysis

# Compute residuals: (data − model at the same data times)
h_model_at_data = tide_model(t, A_fit, w_fit, phi_fit, C_fit)
residuals = h - h_model_at_data
```

```python
#Plot residuals vs time
plt.figure(figsize=(10,4))
plt.axhline(0, color='black', linewidth=1)
plt.plot(t, residuals, '.', markersize=5)
plt.xlabel("Time (hours from start)")
plt.ylabel("Residual (ft)")
plt.title("Residuals of Tide Data After Subtracting Best-Fit Model")
plt.grid(alpha=0.3)
plt.savefig("residuals_plot.pdf")
plt.show()
```

```python
# Histogram of residuals
# “Reasonable” bin width: use Freedman–Diaconis rule
q25, q75 = np.percentile(residuals, [25, 75])
iqr = q75 - q25
bin_width = 2 * iqr * len(residuals) ** (-1/3)
bins = int((residuals.max() - residuals.min()) / bin_width)

plt.figure(figsize=(8,5))
plt.hist(residuals, bins=bins, edgecolor='black')
plt.xlabel("Residual (ft)")
plt.ylabel("Count")
plt.title("Histogram of Tide Model Residuals")
plt.grid(alpha=0.2)
plt.savefig("residuals_histogram.pdf")
plt.show()
```

```python
# Numerical assessment
std_residuals = np.std(residuals, ddof=1)
print("Standard deviation of residuals =", std_residuals)

assumed_sigma = 0.25
intrinsic_scatter = np.sqrt(max(0, std_residuals**2 - assumed_sigma**2))

print("Assumed measurement error = 0.25 ft")
print("Intrinsic scatter in data =", intrinsic_scatter)
```
